```python
# Código completo para graficar tiempos de ejecución en MareNostrum 5
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Helper function
def compute_stats(times):
    avg = np.mean(times)
    err = np.std(times)
    return avg, err

# Datos para ejecución de un solo job (escalado por número de nodos)
single_job_data = {
    'Configuration': [
        '1 Node, 1 Job',
        '2 Nodes, 2 Jobs',
        '4 Nodes, 4 Jobs',
        '8 Nodes, 8 Jobs'
    ],
    'Times': [
        [254.99, 253.70, 270.21, 255.61, 249.64, 263.93, 282.36, 252.06, 275.37, 258.25],
        [249.36, 253.75],
        [247.08, 259.03, 255.37, 277.82],
        [281.13, 269.40, 278.99, 252.78, 281.60, 280.58, 248.07, 251.64]
    ]
}

# Datos para ejecución concurrente (más jobs en los mismos nodos)
concurrent_job_data = {
    'Configuration': [
        '1 Node, 2 Jobs',
        '2 Nodes, 4 Jobs',
        '4 Nodes, 8 Jobs'
    ],
    'Times': [
        [463.29, 496.20],
        [489.24, 494.98, 509.24, 474.60],
        [470.87, 465.02, 491.02, 468.68, 466.38, 493.53, 500.45, 471.14]
    ]
}

# Crear DataFrames y calcular media y desviación estándar
df_single = pd.DataFrame(single_job_data)
df_single[['Average Time (s)', 'Error (s)']] = df_single['Times'].apply(compute_stats).apply(pd.Series)

df_concurrent = pd.DataFrame(concurrent_job_data)
df_concurrent[['Average Time (s)', 'Error (s)']] = df_concurrent['Times'].apply(compute_stats).apply(pd.Series)

# Texto explicativo
textstr = 'Points = μ\nError bars = σ'
props = dict(boxstyle='round', facecolor='lightgray', alpha=0.5)
# Crear gráfico con eje Y limitado de 0 a 500 en ambos subplots

plt.figure(figsize=(14, 6))

# Subplot 1: ejecución individual
plt.subplot(1, 2, 1)
plt.errorbar(df_single['Configuration'], df_single['Average Time (s)'], yerr=df_single['Error (s)'],
             fmt='-o', color='red', ecolor='red', capsize=4)
plt.title('Average Execution Time for Configurations with 112 threads', fontsize=12)
plt.xlabel('Configuration', fontsize=12)
plt.ylabel('Average Execution Time (s)', fontsize=12)
plt.ylim(0, 550)
plt.xticks(rotation=45, ha='right')
plt.grid(True)

# Subplot 2: ejecución concurrente
plt.subplot(1, 2, 2)
plt.errorbar(df_concurrent['Configuration'], df_concurrent['Average Time (s)'], yerr=df_concurrent['Error (s)'],
             fmt='-o', color='blue', ecolor='blue', capsize=4)
plt.title('Average Execution Time for Configurations with 56 threads', fontsize=12)
plt.xlabel('Configuration', fontsize=12)
plt.ylabel('Average Execution Time (s)', fontsize=12)
plt.ylim(0, 550)
plt.xticks(rotation=45, ha='right')
plt.grid(True)

# Añadir caja de explicación
plt.gcf().text(0.95, 0.02, textstr, ha='right', va='bottom', fontsize=8, bbox=props)

plt.tight_layout()
plt.show()

```
